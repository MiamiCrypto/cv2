# Exercise 4 - Annotate Image

```python
# Bring in image from Github

img_weighing_heart = cv2.imread('/content/cv2/WeighingOfTheHeart.png', cv2.IMREAD_COLOR)
plt.figure(figsize=(10,10))
plt.imshow(img_weighing_heart[:, :, ::-1])
```

```python
# Make a copy of the original image
image_annotate = img_weighing_heart.copy()

# Colors to use for shapes
red = (0, 0, 255)
green = (0, 255, 0)
blue = (255, 0, 0)
magenta = (255, 0, 255)
yellow = (0, 255, 255)
cyan = (255, 255, 0)
white = (255, 255, 255)
black = (0, 0, 0)

# Draw a colored rectangles.
image_annotate = cv2.rectangle(image_annotate, (1200, 300), (1480, 760), green, thickness = 3, lineType = cv2.LINE_8)
image_annotate = cv2.rectangle(image_annotate, (500, 300), (700, 500), yellow, thickness = 3, lineType = cv2.LINE_8)

# Draw a red circle.
image_annotate = cv2.circle(image_annotate, (650 , 625), 80, white, thickness = 2, lineType = cv2.LINE_AA)
image_annotate = cv2.circle(image_annotate, (1130 , 625), 60, cyan, thickness = 2, lineType = cv2.LINE_AA)

# Add text to the image.
text = 'Weighing of the Heart'
font_face = cv2.FONT_HERSHEY_DUPLEX
font_scale = 4
font_color = black
font_thickness =14

image_annotate = cv2.putText(image_annotate, text, (200, 125), font_face, font_scale, font_color, font_thickness, cv2.LINE_AA)

# Display the annotated image.
plt.figure(figsize = [12, 12])
plt.imshow(image_annotate[:, :, ::-1]);
```
